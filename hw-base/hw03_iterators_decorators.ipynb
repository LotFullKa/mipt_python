````python
# create shop
shop = CoffeeShop("Moscow")


# dishes
filter_coffee = {
    "id": 1,
    "name": "фильтр-кофе",
    "price": 150,
    "am_left": 3
}

bumble = {
    "id": 2,
    "name": "бамбл",
    "price": 220,
    "am_left": 2
}

almond_latte = {
    "id": 3,
    "name": "айс латте на миндальном",
    "price": 260,
    "am_left": 1
}


shop.add_dish(filter_coffee)
shop.add_dish(bumble)
shop.add_dish(almond_latte)


# ---------- restock_dish success ----------

old_amount = shop.menu[1]["am_left"]

shop.restock_dish(1, am=5)

assert shop.menu[1]["am_left"] == old_amount + 5


# ---------- restock_dish invalid id ----------

error_raised = False

try:
    shop.restock_dish(99, am=5)
except NoDishError:
    error_raised = True

assert error_raised


# ---------- delete_dish success ----------

shop.delete_dish(2)

assert 2 not in shop.menu


# ---------- delete_dish invalid id ----------

error_raised = False

try:
    shop.delete_dish(42)
except NoDishError:
    error_raised = True

assert error_raised


# ---------- form_order success ----------

old_amount = shop.menu[1]["am_left"]

shop.form_order(1)

assert shop.menu[1]["am_left"] == old_amount - 1


# ---------- form_order multiple ids ----------

old1 = shop.menu[1]["am_left"]
old3 = shop.menu[3]["am_left"]

shop.form_order(1, 3)

assert shop.menu[1]["am_left"] == old1 - 1
assert shop.menu[3]["am_left"] == old3 - 1
````
