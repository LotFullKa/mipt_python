#### Какие имена переменных допустимы?

Имена переменных удовлетворяют следующим правилам:

- имя должно содержать только буквы, цифры и знаки нижнего подчеркивания (`_`)
- имя должно начинаться только с буквы или нижнего подчеркивания

**Примечания:**

1. Имена переменных чувствительны к регистру(case-sensetive) - `age`, `Age` и `AGE` являются тремя разными именами
2. Технически у вас есть возможность использовать не только буквы английского алфавита, но и русские, однако, так категорически не рекомендуется делать из-за возможных проблем с кодировкой
3. Советую называть переменные осмысленно, чтобы другой программист смог понять за что именно отвечает та или иная переменная

### NoneType

#### Что такое `None`?

Существует много случаев, когда следует использовать None.

Часто вы хотите выполнить действие, которое может работать либо завершиться неудачно. Используя `None`, вы можете проверить успех действия.

```python
x = None

x == True, x == False, x == None, x == 0, x == ""
```

```python
a = b = 270

a is b, a == b
```

```python
a = None
b = None

a is b, a == b
```

### Циклы: while

```python
x = 17

while x != 1:  # выполняется любое число раз, пока верно условие

    if x % 2 == 0:
        x //= 2
    else:
        x = 3 * x + 1

    print(x)
```

```python
data = [.0, .0, .0, 1.2]

while len(data) > data[-1]:
    last = data[-1]
    data.append(last * last)

    print(data)
```

### Циклы: break, continue, else

#### contnue

```python
# continue заканчивает исполнение текущей ИТЕРАЦИИ
# и сразу же переходит к следущей
```

```python
skip = 7

for n in range(10):
    if n == skip:
        print('*', end=', ')
        continue
    print(n, end=', ')
```

```python
data = [1, 2, 3, 4, 5, 6, 7]

while data:
    value = data.pop()  # удаляем последний элемент
    if 2 <= len(data) <= 4:
        continue
    print(value, end=', ')
```

#### break

```python
# break полностью прерывает исполнение цикла
```

```python
for natural in [1, 2, 3, 4, -1, 20]:
    print(natural)
    if natural < 0:
        print("ERROR")
        break
```

```python
bad = [13, 5, 10]
errors = 0

n = 17
while n > 1:
    print(n, end=', ')
    n = n // 2 if n % 2 == 0 else n * 3 + 1

    if n in bad:
        print('ERROR', end='! ')
        errors += 1
        if errors == len(bad):
            print("FATAL ERROR - BREAK")
            break
```

В Питоне огромное количество разных функций. Если вам нужно сделать какую-то операцию, скорее всего, она уже где-то была написана

 Enumarate

```python
data = [1, 2, 3, 4, 5]

# Новичок
for i in range(len(data)):
    print(i, data[i], end=', ')

print()

# Опытный
for i, value in enumerate(data):
    print(i, value, end=', ')
```

reversed

```python
data = [1, 2, 3, 4, 5]

# Новичок
for value in data[::-1]:
    print(value, end=', ')

print()

# Опытный
for value in reversed(data):
    print(value, end=', ')
```

zip

```python
data = [1, 2, 3, 4, 5]
letters = ['a', 'b', 'c', 'd', 'e']

# Новичок
for i in range(len(data)):
    print(i, (data[i], letters[i]), end=', ')

print()

# Опытный
for value in zip(data, letters):
    print(i, value, end=', ')
```
